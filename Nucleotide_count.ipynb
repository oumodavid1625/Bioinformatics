# **TRANSCRIPTION**
Transcribe DNA and RNA
The replace() method can be used to substitute all occurrences of the DNA base "T" (thymine) with the RNA base "U" (Uracil).
During transcription, RNA polymerase replaces thymine with uracil in the newly synthesized mRNA molecule.

```python
def transcribe_dna_to_rna(dna_sequence):
  """
  Transcribes a DNA sequence into its corresponding RNA sequence.
  Args:
    dna_sequence: A string representing a DNA sequence.
  Returns:
    A string representing the corresponding RNA sequence, or None if the input is invalid.
  """
  if not isinstance(dna_sequence, str):
    raise TypeError("input must be a string")
  if not all(base in "ACGT" for base in dna_sequence.upper()):
    raise valueError("input must be a valid DNA sequence. Must contain only A,C,G,T.")

  rna_sequence = dna_sequence.replace("T", "U")
  return rna_sequence

dna_sequence = "ATGCGATTACGT"
rna_sequence = transcribe_dna_to_rna(dna_sequence)
print(f"DNA sequence: {dna_sequence}")
print(f"RNA sequence: {rna_sequence}")
```

# **Complementing a Strand of DNA**

In DNA strings, symbols 'A' and 'T' are complements of each other, as are 'C' and 'G'.

The DNA molecule is made up of two strands, running in opposite directions.
Each base bonds to a base in the opposite strand. Adenine always bonds with thymine, and cytosine always bonds with guanine; the complement of a base is the base to which it always bonds.
The two strands are twisted together into a long spiral staircase structure called a double helix.

In light of Watson and Crick's model, the bonding of two complementary bases is called a base pair (bp). Therefore, the length of a DNA molecule will commonly be given in bp instead of nt. By complementarity, once we know the order of bases on one strand, we can immediately deduce the sequence of bases in the complementary strand. These bases will run in the opposite order to match the fact that the two strands of DNA run in opposite directions.

```python
def complement_dna(dna_sequence):
  """
  Returns the complementary DNA sequence of the given DNA sequence.
  Args:
      dna_sequence: The DNA sequence string
  Returns:
      The complementary DNA sequence string
  """
  complement_map = {'A': 'T', 'T': 'A', 'C': 'G', 'G': 'C'}
  complementary_sequence = "".join([complement_map[base] for base in dna_sequence])
  return complementary_sequence

dna_strand = "ATGCGCTAG"
complement = complement_dna(dna_strand)
print(f"Original DNA: {dna_strand}")
print(f"Complement: {complement}")
```

# **Complement vs. Reverse Complement in DNA**

The **complement** of a DNA sequence is generated by replacing each base with its Watson-Crick pairing partner:

A ↔ T

C ↔ G

Example:

**Original:** ATGC

**Complement:** TACG

The **reverse complement** is obtained by: First taking the complement (as above). Then reversing the resulting sequence.

Example:

**Original:** ATGC

**Complement:** TACG

**Reverse Complement:** GCAT (complement reversed)


```python
# Complement function
def complement_dna(dna_sequence):
    complement_map = {'A': 'T', 'T': 'A', 'C': 'G', 'G': 'C'}
    return "".join([complement_map[base] for base in dna_sequence])

dna = "ATGC"
print("Original:", dna)                     # ATGC
print("Complement:", complement_dna(dna))    # TACG
```

```python
# Reverse Coplement function
def reverse_complement_dna(dna_sequence):
    complement_map = {'A': 'T', 'T': 'A', 'C': 'G', 'G': 'C'}
    complement = "".join([complement_map[base] for base in dna_sequence])
    return complement[::-1]  # Reverse the complement

dna = "ATGC"
print("Original:", dna)                     # ATGC
print("Reverse Complement:", reverse_complement_dna(dna))  # GCAT
```

**Reverse Complement:**

- Essential for PCR primer design (primers bind to the reverse complement).

- Used in sequence alignment (DNA is double-stranded, so the reverse strand matters).

- Critical for restriction enzyme analysis (enzymes cut the reverse complement).
